### Setting up the data 

```python
import pandas as pd

nba_data = pd.read_csv(r"seasons_clean.csv")
```

```python
# nba_data.info()

# Selecting columns to keep/drop for data set
keep_list = [25,7,0,1,3,5,6,8,9,10,11,12,13,15,17,18,20,21,23,24,26,27,28]
```

```python
# Defining dataset of chosen stats
nba_data2 = nba_data.iloc[:,keep_list]

nba_data2.head(1)
```

```python
# Reserving the 2018 season for "production"
# ***Player name included
nba_data_prod = nba_data2[nba_data2['Season'] == 2018]

# Setting the remainder of data for test and train
nba_data_train_test = nba_data2[nba_data2['Season'] != 2018].drop(columns=["Player"])

# renaming for ease
nba_tt = nba_data_train_test
nba_tt.sort_values()
```

### BUilding X and y dataframe 

```python
# Defining inputs
X = nba_tt.drop(columns=["PER"])

# Defining outputs
y = nba_tt[['PER']].values.reshape(-1,1)

print(X.shape)
print(y.shape)
y.max()
```

### Converting y from array to Categorical dataframe (to One hot encode later) 

```python
df = pd.DataFrame(y)
df.hist(bins=25)
```

```python
# Create the bins in which Data will be held
bins = [-1000, 0, 15, 30, 45, 60, 1000]

# Create the names for the five bins
group_names = ["Very Bad", "Bad", "Avg", "Great", "Excellent","Incredible"]

df[0] = pd.cut(df[0], bins, labels=group_names, include_lowest=True, right=False)
```

```python
df.head(1)
```

```python
y = df[0].to_numpy()
y
```

```python
X
```

### Starting data-modeling process 

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

###### Building Neural net 

```python
from sklearn.preprocessing import StandardScaler

# Create a StandardScater model and fit it to the training data
X_scaler = StandardScaler().fit(X_train)

# Transform the training and testing data using the X_scaler
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

###### One Hot y_train 

```python
from sklearn.preprocessing import LabelEncoder

# Step 1: Label-encode data set
label_encoder = LabelEncoder()
label_encoder.fit(y_train)
encoded_y = label_encoder.transform(y_train)

from tensorflow.keras.utils import to_categorical

# Step 2: One-hot encoding
one_hot_y_train = to_categorical(encoded_y)
one_hot_y_train
```

###### One Hot y_test 

```python
from sklearn.preprocessing import LabelEncoder

# Step 1: Label-encode data set
label_encoder = LabelEncoder()
label_encoder.fit(y_test)
encoded_y = label_encoder.transform(y_test)

from tensorflow.keras.utils import to_categorical

# Step 2: One-hot encoding
one_hot_y_test = to_categorical(encoded_y)
print(type(one_hot_y_test))
```

```python
# first, create a normal neural network with 2 inputs, 6 hidden nodes, and 2 outputs
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

model = Sequential()
model.add(Dense(units=6, activation='relu', input_dim=X.shape[1]))
model.add(Dense(units=6, activation='relu'))
model.add(Dense(units=6, activation='softmax'))

model.summary()
```

```python
# Compile the model
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
# Fit the model to the training data
model.fit(
    X_train_scaled,
    one_hot_y_train,
    epochs=100,
    shuffle=True,
    verbose=2
)
```
