```python
import pandas as pd
from datetime import datetime, timedelta
import numpy as np
```

```python
nba_seasons = "Seasons_Stats.csv"
nba_seasons_df = pd.read_csv(nba_seasons, encoding = "utf-8")
```

```python
nba_seasons_df.head()
```

```python
nba_seasons_df = nba_seasons_df[(nba_seasons_df['Year'] > 2009.0)]
```

```python
nba_seasons_df = nba_seasons_df[(nba_seasons_df['Year'] < 2020.0)]
```

```python
nba_seasons_df.head()
```

```python
nba_seasons_df = nba_seasons_df.drop(columns = ['Unnamed: 0'])
nba_seasons_df.head()
```

```python
nba_seasons_df
```

```python
nba_seasons_df.info()
```

```python
nba_seasons_df['Year'] = pd.to_datetime(nba_seasons_df['Year'], format = '%Y')
nba_seasons_df.head()
```

```python
nba_seasons_df['Year'] = nba_seasons_df['Year'] - pd.offsets.DateOffset(years=1)
nba_seasons_df.head()
```

```python
nba_seasons_df['Season'] = nba_seasons_df['Year'].dt.year
nba_seasons_df.head()
```

```python
#make the df easier to type
df = nba_seasons_df
```

```python
#ppg 
df['PPG'] = df['PTS'].div(df['G'].values)
df['RPG'] = df['TRB'].div(df['G'].values)
df['APG'] = df['AST'].div(df['G'].values)
df['SPG'] = df['STL'].div(df['G'].values)
df['BPG'] = df['BLK'].div(df['G'].values)
df['TOPG'] = df['TOV'].div(df['G'].values)
df['PFPG'] = df['PF'].div(df['G'].values)
df['MPG'] = df['MP'].div(df['G'].values)
df.head()
```

```python
df.info()
```

```python
df = df[['Season', 'Player', 'Pos', 'Age', 'Tm', 'G', 'GS', 'MPG', 'PPG', 'RPG', 'APG', 'SPG', 'BPG', 'TOPG', 'eFG%','FG', 'FGA', 'FG%','3P', '3PA','3P%','FT','FTA','FT%', 'PFPG','PER','TS%','USG%','OWS','DWS','WS','WS/48','OBPM','DBPM','BPM','VORP']]
df.head()
```
