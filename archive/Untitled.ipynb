```python
#import parameters
import pandas as pd
```

```python
#bring in the CSV and convert into a dataframe
nba_data = "all_seasons.csv"
nba_df = pd.read_csv(nba_data, encoding = "utf-8")
```

```python
#check all the columns and datatype
nba_df.info()
```

```python
#drop the first column since it's a copy of the index
nba_df = nba_df.drop(columns = ['Unnamed: 0'])
```

```python
#make sure the unnamed column is dropped
nba_df.head()
```

```python
#separating the year in season by splitting the data through the dash
nba_df[["season 1", "year"]] = nba_df["season"].str.split("-", expand = True)
nba_df.head()
```

```python
nba_df['season 1'] = pd.to_datetime(nba_df['season 1'])
nba_df['season 1']
```

```python
nba_df = nba_df [(nba_df ['season 1'] > '2008-01-01')]
nba_df.head()
```

```python
nba_df = nba_df.drop(columns = ['season 1', 'year'])
```

```python
nba_df.head()
```

```python
nba_df.to_csv ('all_seasons_clean.csv', index = False, header = True)
```
